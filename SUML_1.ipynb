# ex. 4

```python
df_8 = pd.read_csv("DSP_8.csv", sep=',', decimal=',')
```

```python
df_8
```

```python
df_8.info()
```

```python
num_columns = len(df_8.columns)
column_names = df_8.columns.tolist()
print(f"a) Number of columns: {num_columns}\n   Column names: {column_names}")
```

```python
num_rows = len(df_8)
print("b) Number of rows:", num_rows)
```

```python
missing_values = df_8.isnull().sum().sum()
print("c) Total missing values:", missing_values)
```

```python
mean_age_female = df_8[df_8['Sex'] == 'F']['Age'].mean()
std_age_female = df_8[df_8['Sex'] == 'F']['Age'].std()
mean_age_male = df_8[df_8['Sex'] == 'M']['Age'].mean()
std_age_male = df_8[df_8['Sex'] == 'M']['Age'].std()
print(f"d) Mean age for females: {mean_age_female:.2f}, Standard deviation of age for females: {std_age_female:.2f}")
print(f"   Mean age for males: {mean_age_male:.2f}, Standard deviation of age for males: {std_age_male:.2f}")
```

```python
percentage_male = (df_8['Sex'] == 'M').sum() / num_rows * 100
print(f"e) Percentage of males: {percentage_male:.2f}%")
```

```python
num_female_45to50 = ((df_8['Sex'] == 'F') & (df_8['Age'] >= 45) & (df_8['Age'] <= 50)).sum()
print("f) Number of females between the ages of 45 and 50:", num_female_45to50)
```

```python
filtered_df = df_8[df_8['RestingECG'] == 'Normal']
print('g)\n')
plt.figure(figsize=(10, 5))
heatmap = sns.heatmap(filtered_df.corr(numeric_only=True), vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlation Matrix for Individuals with Normal Resting ECG', fontdict={"fontsize":14}, pad=12);
```
